```python
import pandas as pd
import numpy as np
titanic=pd.read_csv(r"C:\Users\new\Downloads\archive (2)\titanic_data.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(x="Sex",hue="Survived",kind="count",data= titanic)
```

```python
group =titanic.groupby(['Pclass','Survived'])
pclass_survived=group.size().unstack()

#heatmap
sns.heatmap(pclass_survived, annot=True, fmt="d")
```

```python
sns.violinplot(x='Sex', y='Age',hue='Survived',data=titanic ,split =True)
```

```python
titanic['Family_Size'] =0
titanic['Family_Size'] = titanic['Parch']+titanic['SibSp']

# adding alone column
titanic['Alone']=0
titanic.loc[titanic.Family_Size==0,'Alone']=1

sns.violinplot(x ='Family_Size', y ='Survived',data =titanic)
sns.barplot(x ='Alone' ,y ='Survived',data =titanic)



```

```python
# division onthe basis of fair 
titanic['Fair_Range']=pd.qcut(titanic['Fare'],4)
sns.barplot(x='Fair_Range', y= 'Survived',data = titanic)
                              
```

```python
#countplot
sns.catplot(x='Embarked',hue= 'Survived',kind='count',col='Pclass',data =titanic)
```
